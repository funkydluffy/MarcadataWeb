```python
import matplotlib.pyplot as plt
import numpy as np

# Configuración del tamaño de la imagen
fig, ax = plt.subplots(figsize=(10, 10))  # Creamos una figura de 10x10 pulgadas

# Parámetros de diseño
size = 1  # Tamaño del radio de cada hexágono
cols = 10  # Número de columnas de hexágonos en la imagen
rows = 10  # Número de filas de hexágonos en la imagen

# Creación de los hexágonos
for row in range(rows):  # Iteramos sobre cada fila
    for col in range(cols):  # Iteramos sobre cada columna
        # Calculamos el desplazamiento en el eje X para la posición del hexágono
        x_offset = col * 1.5 * size
        
        # Calculamos el desplazamiento en el eje Y para la posición del hexágono
        y_offset = row * np.sqrt(3) * size
        
        # Si la columna es impar, ajustamos el hexágono hacia abajo para crear el patrón de colmena
        if col % 2 == 1:
            y_offset += np.sqrt(3) / 2 * size
        
        # Definimos los vértices del hexágono utilizando las coordenadas polares
        hexagon = plt.Polygon(
            [[np.cos(np.pi / 3 * i) * size + x_offset, np.sin(np.pi / 3 * i) * size + y_offset] 
             for i in range(6)],  # Calculamos las coordenadas de los 6 vértices del hexágono
            closed=True  # Cerramos el polígono para formar un hexágono
        )
        
        # Agregamos el hexágono al gráfico
        ax.add_patch(hexagon)

# Personalización del gráfico
ax.set_aspect('equal')  # Asegura que los hexágonos se vean proporcionados (aspecto 1:1)
ax.axis('off')  # Desactiva los ejes para que solo se vea el patrón de hexágonos

# Guardamos la imagen generada como un archivo PNG
plt.savefig("hexagon_background.png", bbox_inches='tight', pad_inches=0)

# Mostramos el gráfico en pantalla
plt.show()
```

```python
# Configuración del tamaño de la imagen y el color de fondo
fig, ax = plt.subplots(figsize=(10, 10))
fig.patch.set_facecolor('white')  # Fondo blanco para la figura

# Parámetros de diseño
size = 1  # Tamaño del radio de cada hexágono
cols = 10  # Número de columnas de hexágonos
rows = 10  # Número de filas de hexágonos

# Función para crear un hexágono con efecto 3D
def draw_hexagon(x_offset, y_offset, color='white', edge_color='lightgrey', shadow_color='grey'):
    # Vértices del hexágono usando coordenadas polares
    hexagon_points = [[np.cos(np.pi / 3 * i) * size + x_offset, np.sin(np.pi / 3 * i) * size + y_offset] for i in range(6)]
    
    # Dibujamos la sombra para crear el efecto 3D
    shadow_points = [[np.cos(np.pi / 3 * i) * size * 1.05 + x_offset, np.sin(np.pi / 3 * i) * size * 1.05 + y_offset] for i in range(6)]
    shadow_hexagon = plt.Polygon(shadow_points, closed=True, color=shadow_color, alpha=0.2)  # Sombra con menor opacidad
    ax.add_patch(shadow_hexagon)
    
    # Dibujamos el hexágono principal encima de la sombra
    hexagon = plt.Polygon(hexagon_points, closed=True, edgecolor=edge_color, facecolor=color, linewidth=1.5)
    ax.add_patch(hexagon)

# Crear la cuadrícula de hexágonos
for row in range(rows):
    for col in range(cols):
        # Calculamos el desplazamiento en X e Y para colocar el hexágono
        x_offset = col * 1.5 * size
        y_offset = row * np.sqrt(3) * size
        
        # Desplazamos las columnas impares para crear el patrón de colmena
        if col % 2 == 1:
            y_offset += np.sqrt(3) / 2 * size
        
        # Dibujamos el hexágono con un borde gris claro y sombra gris
        draw_hexagon(x_offset, y_offset, color='white', edge_color='lightgrey', shadow_color='grey')

# Personalización del gráfico
ax.set_aspect('equal')  # Mantiene la proporción 1:1
ax.axis('off')  # Oculta los ejes para que solo se vea el fondo de hexágonos

# Guardamos la imagen generada como un archivo PNG
plt.savefig("hexagon_3d_background.png", bbox_inches='tight', pad_inches=0, dpi=300)

# Mostramos el gráfico en pantalla
plt.show()
```
